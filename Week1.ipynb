### 5 Handle missing values 

####  5.1   Finding median of Age and mode of Embarked


```python
median1=data['Age'].median()
print(median1)
mode1=data['Embarked'].mode()
print("mode:",mode1)
```

#### 5.2   Fill missing values with median for 'Age' column , mode for 'Embarked' column


```python
data['Age'].fillna(value=median1,inplace=True)
data['Embarked'].fillna(value=mode1,inplace=True)
data
```

### 6 Drop rows with missing values
#### (Drop rows with missing values in 'Cabin' column)



```python
data.dropna(subset=['Cabin'],inplace=True)
data
```

### 9 Remove outliers based on ‘Fare’ column


```python
Q1 = data['Fare'].quantile(0.25)
Q3 = data['Fare'].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Remove outliers
data = data[(data['Fare'] >= lower_bound) & (data['Fare'] <= upper_bound)]
data
```
