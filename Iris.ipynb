### Importing Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load the Iris dataset

```python

iris = sns.load_dataset('iris')
```

### Scatter plot for 'sepal length (cm)' and 'sepal width (cm)'

```python

plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris)
plt.title('Scatter plot of Sepal Length vs Sepal Width')
plt.show()
```

### Histogram for 'petal length (cm)'

```python

plt.figure(figsize=(8, 6))
sns.histplot(iris['petal_length'], bins=20, kde=True)
plt.title('Histogram of Petal Length')
plt.show()
```

### Box plot for 'species' and 'petal width (cm)'

```python

plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_width', data=iris)
plt.title('Box plot of Petal Width by Species')
plt.show()
```
